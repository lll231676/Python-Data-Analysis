# 1 评估数据

### 1.1评估数据整齐度

```python
original_data.sample(10)
```

### 从抽样的10行数据中，数据符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”
### 因此不存在结构性问题

### 1.2评估数据干净度

```python
original_data.info()
```

 ### 从输出信息来看 一共有51909条观察值，而Description、CustomerID变量存在缺失值 
 ### 因为他们的Non-Null非空缺值数量少于我们的总行数 此为INvoiceDate应该为日期类型 
 ### CustomerID应该为字符串类型 而不是浮点数类型

### 1.3评估缺失数据

### 筛选出Description值缺失的那些行

```python
original_data[original_data["Description"].isnull()]
```

### 有1454条数据缺失Description 其中在显示的数据中UnitPrice都为0.0 为了验证
### 是否有存在不为0的数据 

```python
original_data[(original_data["Description"].isnull()) &(original_data["UnitPrice"]!=0.0)]
```

### 这些UnitPrice为0的数据 可以在后续都删除掉

```python
original_data[original_data["CustomerID"].isnull()]
```

### CustomerID表示用户编号 不是分析畅销产品的必须变量 
### 因此缺失用户编号的那些数据仍然具备分析效力

### 1.4评估重复数据

### 根据数据变量的定义 我们对于唯一存在的变量需要进行重复数据的删除
### 如InvoiceNo(订单编号)、StockCode(货物编号)和CustmoerID(用户id)
### 但是在电商项目中 同一个订单号可以购买相同的商品 因此可以有相同的商品编号
### 同时同一个用户可以下单不同的商品 这些商品也可以隶属于一个订单 

### 因此，不需要做重复数据的评估

### 1.5评估不一致数据


### 不一致的数据可以存在于Country(国家)中，我们要看是否有多个字符串指代同一个国家的情况
### 如"china","CHINA"

```python
original_data["Country"].value_counts()
```

### 没有发现不一致数据 如"UK","United Kingdom"都是英国的意思

### 1.6评估无效或错误数据

### 可以通过DataFrame的describe方法进行快速了解

```python
original_data.describe()
```

### 我们发现数量 和 单价 出现了负数 这会对后续的数据分析产生应该
### 因此 我们把负数的列筛选出来 进一步评估其意义

```python
original_data[original_data["Quantity"] < 0]
```

### 我们可以看到在输出的结果中数量为负数的这些行对应的订单编号都是以C开头的
### 订单编号以C开头的都是已经取消的订单 不妨让我们猜测以下是否取消的订单的数量都是负数呢

### 让我们显示一下编号为C开头但是数量不为负数的这些行 来验证我们的想法

```python
original_data[(original_data["Quantity"] < 0) & (original_data["InvoiceNo"].str[0]!='C')]
```

### 又上述条件 我们可以发现数量Quantity全为0的订单ID并不全部都是C开头
### 但是我们可以发现这个价格UnitPrice全为0 我们可以提出下一个猜测
### 是否数量Quantity全为负数和订单ID并不全部都是C开头的情况 单价UnitPrice全为0呢

```python
original_data[(original_data["Quantity"] < 0) & (original_data["InvoiceNo"].str[0]!='C') & (original_data["UnitPrice"]!=0.0)]
```

### 由于输出结果可视 我们知道数量Quantity小于0的数据满足以下条件之一：
### 1订单编号以C开头 为取消的订单
### 2单价都为0  
### 这些数据都不具有分析意义 反而会影响数据分析 因为考虑删除这些数据



### 接下来筛选出单价UnitPrice小于0的数据

```python
original_data[(original_data["UnitPrice"] < 0)]
```


### 我们可以发现这些单价小于01的数据中的秒速都是坏账调整 不属于实际商品交易数据 因此可以删除

# (2)清理数据

### 根据我们前面对数据的评估 我们需要清理的数据如下：
 - 把InvoiceDate的类型转换为日期时间
 - 把CustomerID的类型转换为字符串
 - 把Description的缺失值变量删除
 - 把Quantity中的负数值删除
 - 把UnitPrice中的负数值删除

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data["InvoiceDate"]
```

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype(str)
cleaned_data["CustomerID"]
```

### 我们想把编号后面的.0给删除

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].str.slice(0,-2)
cleaned_data["CustomerID"]
```

```python
cleaned_data.dropna(subset=["Description"],inplace=True)
```

### 验证表中Decription是否还有空值

```python
cleaned_data["Description"].isnull().sum()
```

### 如果我们需要对"CHINA","cn"都替换为"china"我们可以使用以下语法
### cleaned_data["Country"] = cleaned_data["Country"].replace({"CHINA":"china","cn":"china"})
### 接着可以用len（）进行检查




### 删除负数操作如下

```python
cleaned_data =cleaned_data[cleaned_data["Quantity"] >= 0]
```

```python
len(cleaned_data[cleaned_data["Quantity"] <0])
```

```python
cleaned_data =cleaned_data[cleaned_data["UnitPrice"] >= 0]
```
